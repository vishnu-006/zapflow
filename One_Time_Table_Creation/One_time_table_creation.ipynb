```python
import json
```

```python
raw_path = dbutils.notebook.entry_point.getDbutils().notebook().getContext().notebookPath().get()
dir_name = "/".join(raw_path.split("/")[:-2])
config_file_path = f"/Workspace{dir_name}/process_config.json"
# dbutils.widgets.removeAll()
dbutils.widgets.text("config_file_path", config_file_path)
temp = dbutils.widgets.get("config_file_path")
# print(temp)
```

```python
with  open(dbutils.widgets.get("config_file_path"), "r") as file:
# with  open("/Workspace/Repos/------/zapflow/process_config.json", "r") as file:
  config = file.read()
config = json.loads(config)
# print(config)
# print(type(config))
```

```python
sql_script = f"""
CREATE OR REPLACE TABLE {config['raw_catalog_name']}.{config['raw_schema_name']}.customers(
    
    customer_id           STRING   COMMENT 'Unique identifier for each customer',
    first_name            STRING  COMMENT 'Customer''s first name',
    last_name             STRING  COMMENT 'Customer''s last name',
    email                 STRING  COMMENT 'Customer''s email address',
    phone                 STRING  COMMENT 'Customer''s phone number',
    city                  STRING  COMMENT 'Customer''s city of residence',
    pincode               INTEGER     COMMENT 'Customer''s postal code',
    customer_segment      STRING  COMMENT 'Segment/category of the customer',
    acquisition_channel   STRING  COMMENT 'Channel through which customer was acquired',
    registration_date     DATE    COMMENT 'Date of customer registration',      
    is_active             STRING  COMMENT 'Indicates if customer is active',
    _data_quality_issue   STRING  COMMENT 'Data quality issue flag or description',
    ingestion_time        TIMESTAMP COMMENT 'Timestamp when data was ingested',
    source_file           STRING  COMMENT 'Source file path'
    
    )

"""
spark.sql(sql_script)
```

```python
sql_script = f"""
CREATE OR REPLACE TABLE {config['raw_catalog_name']}.{config['raw_schema_name']}.dark_stores(
    store_id               STRING      COMMENT 'Unique identifier for each store',
    store_name             STRING      COMMENT 'Name of the store',
    city                   STRING      COMMENT 'City where the store is located',
    zone                   STRING      COMMENT 'Zone of the city',
    address                STRING      COMMENT 'Store address',
    pincode                INTEGER     COMMENT 'Store postal code',
    capacity_sqft          INTEGER     COMMENT 'Store capacity in square feet',
    max_orders_per_day     INTEGER     COMMENT 'Maximum orders the store can handle per day',
    operational_since      DATE        COMMENT 'Date when the store became operational',
    store_status           STRING      COMMENT 'Current status of the store',
    delivery_radius_km     DOUBLE      COMMENT 'Delivery radius in kilometers',
    _data_quality_issue    STRING      COMMENT 'Data quality issue flag or description',
    ingestion_time         TIMESTAMP   COMMENT 'Timestamp when data was ingested',
    source_file            STRING      COMMENT 'Source file path'
)
"""
spark.sql(sql_script)
```

```python
sql_script = f"""
CREATE OR REPLACE TABLE {config['raw_catalog_name']}.{config['raw_schema_name']}.deliveries(
    delivery_id           STRING        COMMENT 'Unique identifier for each delivery',
    order_id              STRING        COMMENT 'Unique identifier for each order',
    delivery_partner_id   STRING        COMMENT 'Unique identifier for delivery partner',
    store_id              STRING        COMMENT 'Store from which delivery is made',
    pickup_timestamp      TIMESTAMP     COMMENT 'Timestamp when order was picked up',
    delivered_timestamp   TIMESTAMP     COMMENT 'Timestamp when order was delivered',
    delivery_minutes      DOUBLE        COMMENT 'Total delivery time in minutes',
    is_sla_breached       STRING        COMMENT 'Indicates if SLA was breached',
    delivery_status       STRING        COMMENT 'Status of the delivery',
    _data_quality_issue   STRING        COMMENT 'Data quality issue flag or description',
    ingestion_time        TIMESTAMP     COMMENT 'Timestamp when data was ingested',
    source_file           STRING        COMMENT 'Source file path'
)
"""
spark.sql(sql_script)
```

```python
sql_script = f"""
CREATE OR REPLACE TABLE {config['raw_catalog_name']}.{config['raw_schema_name']}.inventory(
    inventory_id         STRING      COMMENT 'Unique identifier for inventory record',
    store_id             STRING      COMMENT 'Store identifier',
    product_name         STRING      COMMENT 'Name of the product',
    category             STRING      COMMENT 'Product category',
    snapshot_date        DATE        COMMENT 'Date of inventory snapshot',
    units_available      INTEGER     COMMENT 'Units available in inventory',
    units_sold           INTEGER     COMMENT 'Units sold',
    reorder_level        INTEGER     COMMENT 'Reorder threshold level',
    is_out_of_stock      STRING      COMMENT 'Indicates if product is out of stock',
    is_weekend           STRING      COMMENT 'Indicates if snapshot date is a weekend',
    _data_quality_issue  STRING      COMMENT 'Data quality issue flag or description',
    ingestion_time       TIMESTAMP   COMMENT 'Timestamp when data was ingested',
    source_file          STRING      COMMENT 'Source file path'
)
"""
spark.sql(sql_script)
```
