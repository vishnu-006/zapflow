```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DateType
from pyspark.sql.functions import col, current_timestamp
```

```python
def read_and_write(Schema, Table_Name, Source_Path):
    df = spark.read.format("csv").option("header", "true").schema(Schema).load(Source_Path).withColumn("ingestion_time",current_timestamp()).withColumn("source_file", col("_metadata.file_path"))

    df.write.mode("append").saveAsTable(Table_Name)
```
